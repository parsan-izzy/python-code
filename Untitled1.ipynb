```python
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine

# Replace 'your_database_name' with your actual database name
connection_string = 'mssql+pyodbc://DESKTOP-G22B874/AdventureWorks2022?trusted_connection=yes&driver=ODBC+Driver+17+for+SQL+Server'
engine = create_engine(connection_string)

# Your SQL query
query = """
SELECT 
    YEAR(wo.EndDate) AS ProductionYear,
    ISNULL(cat.Name, 'NONE') AS ProductCategory,
    SUM(wo.StockedQty) AS QuantityProduced
FROM 
    Production.WorkOrder wo
JOIN 
    Production.Product pr ON wo.ProductID = pr.ProductID
LEFT JOIN 
    Production.ProductSubcategory sub ON pr.ProductSubcategoryID = sub.ProductSubcategoryID
LEFT JOIN 
    Production.ProductCategory cat ON sub.ProductCategoryID = cat.ProductCategoryID
GROUP BY 
    YEAR(wo.EndDate),
    ISNULL(cat.Name, 'NONE')
"""

# Execute the SQL query and read data into a DataFrame
df = pd.read_sql(query, con=engine)

# Pivot the DataFrame for better visualization
pivot_df = df.pivot(index='ProductionYear', columns='ProductCategory', values='QuantityProduced')

# Plotting
pivot_df.plot(kind='bar', stacked=True)
plt.title('Total Production Quantity by Year and Product Category')
plt.xlabel('Production Year')
plt.ylabel('Quantity Produced')
plt.legend(title='Product Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Close the database connection
engine.dispose()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sqlalchemy import create_engine

# Replace 'your_database_connection_string' with your actual connection string
engine = create_engine('mssql://DESKTOP-G22B874/AdventureWorks2022?trusted_connection=yes&driver=ODBC+Driver+17+for+SQL+Server')
connection = engine.connect()

# Replace the query with your actual SQL query
query = """
  SELECT wo.ProductID
  ,pr.Name As 'ProductName'
  ,ISNULL(cat.Name, 'Unknown') AS 'ProductCategory'
  ,	Format(EndDate, 'MM-yyyy') AS 'ProductionMonthYear'
  ,SUM(wo.[StockedQty]) AS 'QuantityProduced'
 
 FROM Production.WorkOrder wo LEFT OUTER JOIN Production.Product pr 
  ON wo.ProductID = pr.ProductID
  LEFT OUTER JOIN Production.ProductSubcategory sub 
  ON pr.ProductSubcategoryID = sub.ProductSubcategoryID
  LEFT OUTER JOIN Production.ProductCategory cat 
  ON sub.ProductCategoryID = cat.ProductCategoryID
--WHERE

 GROUP BY wo.productID
 , pr.Name
 , ISNULL(cat.Name, 'Unknown')
 , Format(EndDate, 'MM-yyyy') 
 , wo.EndDate


ORDER BY 
year(Enddate), Month(Enddate)
"""

# Load data into a pandas DataFrame
df = pd.read_sql(query, con=connection)

# Pivot the data to create a DataFrame suitable for plotting
df_pivot = df.pivot_table(values='QuantityProduced', index='ProductionMonthYear', columns='ProductCategory', aggfunc='sum')

# Plotting the stacked bar chart
ax = df_pivot.plot(kind='bar', stacked=True, figsize=(12, 6))
ax.set_xlabel('Production Month Year')
ax.set_ylabel('Quantity Produced')
ax.set_title('Stacked Bar Chart of Quantity Produced by Product Category and Production Month')

# Show the plot
plt.show()
```
